### Load Data from Excel Files

```python
import pandas as pd
```

#### Loading Natural Gas Price Data

```python
# Load Natural Gas Price data
ng_price_df = pd.read_excel('/content/Natural_Gas_Price.xlsx')

print('Natural Gas Price DataFrame - First row:')
display(ng_price_df.iloc[0])

print('\nNatural Gas Price DataFrame - Last row:')
display(ng_price_df.iloc[-1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the line plot
sns.lineplot(x='Date', y='Settlement', data=ng_price_df)

# Add title and labels for clarity
plt.title('Natural Gas Price (Settlement) Over Time')
plt.xlabel('Date')
plt.ylabel('Settlement Price')

# Rotate x-axis labels for better readability if dates overlap
plt.xticks(rotation=45)

# Add a grid for better visualization
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

#### Loading Coal Price Data

```python
# Load Coal Price data
coal_price_df = pd.read_excel('/content/Coal_Price.xlsx')

print('Coal Price DataFrame - First row:')
display(coal_price_df.iloc[0])

print('\nCoal Price DataFrame - Last row:')
display(coal_price_df.iloc[-1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the line plot
sns.lineplot(x='Date', y='Settlement', data=coal_price_df)

# Add title and labels for clarity
plt.title('Coal Price (Settlement) Over Time')
plt.xlabel('Date')
plt.ylabel('Settlement Price')

# Rotate x-axis labels for better readability if dates overlap
plt.xticks(rotation=45)

# Add a grid for better visualization
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

#### Calculating Coal Returns

```python
# Ensure 'Date' column is in datetime format and sort the DataFrame by date
coal_price_df['Date'] = pd.to_datetime(coal_price_df['Date'])
coal_price_df = coal_price_df.sort_values(by='Date').reset_index(drop=True)

# Calculate Coal Return as the percentage change in Settlement price
coal_price_df['Coal_Return'] = coal_price_df['Settlement'].pct_change()

print('Coal Price DataFrame with Returns - First 5 rows:')
display(coal_price_df.head())

print('\nCoal Price DataFrame with Returns - Last 5 rows:')
display(coal_price_df.tail())

print('\nSummary statistics for Coal_Return:')
display(coal_price_df['Coal_Return'].describe())
```

```python
print('Full Summary statistics for Coal_Return:')
display(coal_price_df['Coal_Return'].describe())
```

#### Calculating Natural Gas Returns

```python
# Ensure 'Date' column is in datetime format and sort the DataFrame by date
ng_price_df['Date'] = pd.to_datetime(ng_price_df['Date'])
ng_price_df = ng_price_df.sort_values(by='Date').reset_index(drop=True)

# Calculate Natural Gas Return as the percentage change in Settlement price
ng_price_df['NatGas_Return'] = ng_price_df['Settlement'].pct_change()

print('Natural Gas Price DataFrame with Returns - First 5 rows:')
display(ng_price_df.head())

print('\nNatural Gas Price DataFrame with Returns - Last 5 rows:')
display(ng_price_df.tail())

print('\nSummary statistics for NatGas_Return:')
display(ng_price_df['NatGas_Return'].describe())
```

#### Combined Plots: Coal and Natural Gas Prices and Returns

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 12)) # Adjust figure size as needed
fig.suptitle('Coal and Natural Gas Prices and Returns Over Time', fontsize=16)

# Plot 1: Coal Price (Settlement) Over Time
sns.lineplot(x='Date', y='Settlement', data=coal_price_df, ax=axes[0, 0])
axes[0, 0].set_title('Coal Price (Settlement) Over Time')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Settlement Price')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, linestyle='--', alpha=0.7)

# Plot 2: Coal Return Over Time
sns.lineplot(x='Date', y='Coal_Return', data=coal_price_df, ax=axes[0, 1])
axes[0, 1].set_title('Coal Return Over Time')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Coal Return')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, linestyle='--', alpha=0.7)

# Plot 3: Natural Gas Price (Settlement) Over Time
sns.lineplot(x='Date', y='Settlement', data=ng_price_df, ax=axes[1, 0])
axes[1, 0].set_title('Natural Gas Price (Settlement) Over Time')
axes[1, 0].set_xlabel('Date')
axes[1, 0].set_ylabel('Settlement Price')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, linestyle='--', alpha=0.7)

# Plot 4: Natural Gas Return Over Time
sns.lineplot(x='Date', y='NatGas_Return', data=ng_price_df, ax=axes[1, 1])
axes[1, 1].set_title('Natural Gas Return Over Time')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Natural Gas Return')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Calculating 100-day Moving Average of Volatility for Coal Returns

```python
# Calculate the 100-day rolling standard deviation of 'Coal_Return'
# This represents the 100-day moving average of volatility
coal_price_df['Coal_Vol_MA100'] = coal_price_df['Coal_Return'].rolling(window=100).std()

print('Coal Price DataFrame with 100-day Volatility MA - First 5 rows:')
display(coal_price_df.head())

print('\nCoal Price DataFrame with 100-day Volatility MA - Last 5 rows:')
display(coal_price_df.tail())

print('\nSummary statistics for Coal_Vol_MA100:')
display(coal_price_df['Coal_Vol_MA100'].describe())
```

#### Plotting 100-day Moving Average of Volatility for Coal Returns Over Time

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the line plot for Coal_Vol_MA100
sns.lineplot(x='Date', y='Coal_Vol_MA100', data=coal_price_df)

# Add title and labels for clarity
plt.title('Coal Volatility (100-day Moving Average) Over Time')
plt.xlabel('Date')
plt.ylabel('100-day Rolling Std Dev of Returns')

# Rotate x-axis labels for better readability if dates overlap
plt.xticks(rotation=45)

# Add a grid for better visualization
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the 100-day rolling standard deviation of 'NatGas_Return'
# This represents the 100-day moving average of volatility
ng_price_df['NatGas_Vol_MA100'] = ng_price_df['NatGas_Return'].rolling(window=100).std()

print('Natural Gas Price DataFrame with 100-day Volatility MA - First 5 rows:')
display(ng_price_df.head())

print('\nNatural Gas Price DataFrame with 100-day Volatility MA - Last 5 rows:')
display(ng_price_df.tail())

print('\nSummary statistics for NatGas_Vol_MA100:')
display(ng_price_df['NatGas_Vol_MA100'].describe())
```
