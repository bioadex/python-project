# Create lollipop Chart!!! 

```python
# create a lolipop chat     
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
df_quality = df['quality'].value_counts().sort_index()

# Create a figure and axis
plt.figure()

# Plot the stems (lines)
plt.stem(df_quality.index, df_quality.values)

# Add markers at the top of each line
plt.scatter(df_quality.index, df_quality.values, color='red', zorder=3)

# Add titles and labels
plt.title('Ratio of Quality in Wine')
plt.xlabel('Wine Quality')
plt.ylabel('Number of Wines')

# Show the plot
plt.show()
```

# Create Plot Layout Chart 

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.transform import cumsum
from math import pi
import pandas as pd
from bokeh.palettes import Category20c
from bokeh.models import ColumnDataSource

# Assuming df is your DataFrame
df_quality = df['quality'].value_counts()

# Output file
output_file("wine_quality_plots.html")

# ----- Plot 1: Horizontal Bar Plot -----
bar_fig = figure(title="Ratio of Quality in Wine", 
                 x_axis_label='Number of Wines', 
                 y_axis_label='Quality', 
                 min_height=350, min_width=600)

bar_fig.hbar(y=df_quality.index.astype(str), right=df_quality.values, height=0.4, color='skyblue')

# ----- Plot 2: Donut Chart -----
# Prepare data for the donut chart
df_donut = df_quality.reset_index(name='value').rename(columns={'index': 'quality'})
df_donut['angle'] = df_donut['value'] / df_donut['value'].sum() * 2 * pi
df_donut['color'] = Category20c[len(df_donut)]

donut_fig = figure(title="Donut Chart: Quality of Wine", 
                   min_height=350, min_width=350, 
                   toolbar_location=None, tools="hover", tooltips="@quality: @value", x_range=(-0.5, 1.0))

donut_fig.wedge(x=0, y=1, radius=0.4, start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
                line_color="white", fill_color='color', legend_field='quality', source=df_donut)

# Add a white circle at the center to make it a donut chart
donut_fig.circle(x=0, y=1, radius=0.2, fill_color="white", line_color="white")

# ----- Plot 3: Pie Chart -----
pie_fig = figure(title="Pie Chart: Quality of Wine", 
                 min_height=350, min_width=350, 
                 toolbar_location=None, tools="hover", tooltips="@quality: @value", x_range=(-0.5, 1.0))

pie_fig.wedge(x=0, y=1, radius=0.4, start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
              line_color="white", fill_color='color', legend_field='quality', source=df_donut)

# Remove axis visibility for donut and pie charts
for fig in [donut_fig, pie_fig]:
    fig.axis.visible = False
    fig.grid.grid_line_color = None

# ----- Layout -----
layout = gridplot([[bar_fig], [donut_fig, pie_fig]])

# Show the layout
show(layout)
```
