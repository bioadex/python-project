```python
# Draw Contour Image 

import matplotlib
import numpy as np 
import matplotlib.cm as cm 
import matplotlib.pyplot as plt 

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 -Z2) * 2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

a = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]
A, B = np.meshgrid(a, np.linspace(-3, 3, 128))
C = (1 - A/2 + A**5 + B**3) * np.exp(-A**2 - B**2)

# plot:
fig, ax = plt.subplots(2, 4, figsize=(12,5))
ax[0,0].set_title('hexbin')
ax[0,0].hexbin(x, y, gridsize=20)
ax[0,0].set(xlim=(-2, 2), ylim=(-3, 3))

ax[0,1].set_title('hist2D')
ax[0,1].hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))
ax[0,1].set(xlim=(-2, 2), ylim=(-3, 3))

ax[0,2].set_title('ecdf x')
ax[0,2].ecdf(x)
# ax[0,2].set(xlim=(-2, 2), ylim=(-3, 3))

ax[0,3].set_title('imshow')
ax[0,3].imshow(Z, origin='lower')

ax[1,0].set_title('pcolormesh')
ax[1,0].pcolormesh(A, B, C, vmin=-0.5, vmax=1.0)

# make data
D, E = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
F = (1 - D/2 + D**5 + E**3) * np.exp(-D**2 - E**2)
levels = np.linspace(F.min(), F.max(), 7)

ax[1,1].set_title('contourf')
ax[1,1].contourf(D, E, F, levels=levels)
```
