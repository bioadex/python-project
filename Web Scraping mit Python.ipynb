```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re 
```

```python
url="https://my.price-it.eu/de/"      #Open url link 
html=urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a')
print(links)
```

```python
links = soup.find_all('a', href=True)
for link in links:
    print(link['href'])
```

```python
allrows = soup.find_all('tr')
print(allrows)
```

```python
data = []
allrows = soup.find_all('tr')
for row in allrows:
    row_list = row.find_all("td")
    dataRow = []
    for cell in row_list:
        dataRow.append(cell.text)
        data.append(dataRow)
print(data)
```
