```python
class Person :
    def __init__(self,n,g,a):
        self.name=n
        self.gender=g
        self.age=a
    def talk(self):
        print("Hi I'm ", self.name, self.age)
    def vote(self):
        if self.age<18 :
            print("I am not eligible to vote")
        else :
            print("I am eligible to vote")
obj1= Person("Sam","Male",22)
obj2= Person("Jenny","Female",16)

obj1.talk()
obj1.vote()

obj2.talk()
obj2.vote()
```

### Python OOPs Concepts: Object Oriented Programming Concept? ####

```python
class apartment():
    
    def __init__(self, year, price):
       self.year = year
       self.price = price
    def getPrice(self):
        print("The Price is: ", self.price)
    def setPrice(self,price):
        self._price=price
        print(self._price)
        
Duplex = apartment(1998,90000)
Skyscrapper = apartment(2020,1000000)
Duplex.setPrice = apartment(2010,150000)

Duplex.getPrice()
Skyscrapper.getPrice()
Duplex.getPrice()
```

```python
from threading import *
def show() : 
    print("This is a child thread")
t= Thread(target= show() )
t.start()
print("This is a parent thread")
```

```python
from threading import *
class MyThread(Thread):
    for i in range(5):
        print("\nThis is a child class")
t= MyThread()
t.start()
for i in range(5):
    print("\nThis is the main thread")
```

```python
from threading import *
class Demo:
    def show(self):
        for i in range(5):
            print("\nThis is the child thread")
obj=Demo()
t=Thread(target=obj.show())
t.start()
for i in range(5):
    print("\nThis is the parent thread")
```

```python
#### MULTITHREEADING ####
```

```python
from threading import *
import time
class Demo:
    def num(self):
        for i in range(1,6):
            print("The number is", i)
            time.sleep(1)
    def double(self):
        for i in range(1,6):
            print("The double of the nunber is", 2*i)
            time.sleep(1)
    def square(self):
        for i in range (1,6):
            print("The square of the number is", i*i)
            time.sleep(1)
obj=Demo()
t1= Thread(target=obj.num)
t2= Thread(target=obj.double)
t3=Thread(target=obj.square)

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("This is the main thread")
```
